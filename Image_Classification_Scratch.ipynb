```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# fix random seed for reproducibility
np.random.seed(43)
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D,Activation
from keras.optimizers import Adam ,RMSprop
from sklearn.model_selection import train_test_split
from keras import  backend as K
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img_rows,img_cols = 28,28 #height and width of the pixel
num_classes = 10 #Total number of classes

```

```python
import keras

def data_prep(raw):
    # One-hot encode labels using the updated function
    train_y = keras.utils.to_categorical(raw['label'], num_classes)

    num_images = raw.shape[0]
    x_as_array = raw.iloc[:, 1:].values  # Extract pixel values safely
    x_shaped_array = x_as_array.reshape(num_images, img_rows, img_cols, 1)

    # Normalize the image data
    train_x = x_shaped_array.astype('float32') / 255.0

    return train_x, train_y

# Example DataFrame (Replace with your actual dataset)
# train = pd.read_csv('your_dataset.csv')

train_x, train_y = data_prep(train)
```

```python
#Building the model

from keras.utils import to_categorical
Y_train = train["label"]
X_train = train.drop(columns=['label'], axis=1)
X_train = X_train / 255.0
test = test / 255.0
X_train = X_train.values.reshape(-1,28,28,1)
test = test.values.reshape(-1,28,28,1)
Y_train = to_categorical(Y_train, num_classes = 10)
```

```python
random_seed = 2
X_train,X_test,Y_train,Y_val = train_test_split(X_train,Y_train,test_size=0.1,random_state=random_seed)
```

```python
model = Sequential()
model.add(Dense(64,input_shape=(28,28,1)))
model.add(Activation('relu'))
model.add(Flatten()) # Converts in to vector
model.add(Dense(10,activation = 'softmax'))
model.summary()
```

```python
model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adam(),metrics=['accuracy'])
model.fit(X_train,Y_train,batch_size=128,epochs=10,validation_data=(X_test,Y_val))

```

```python
#Now we Split the tranning data randomly with seed=42

X_train,X_val,Y_train,Y_val = train_test_split(train_x,train_y,test_size=0.1,random_state=2)

```

```python
model = Sequential()
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same',
                 activation ='relu', input_shape = (28,28,1)))
model.add(Conv2D(filters = 32,kernel_size = (5,5),padding = 'same',activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',
                 activation ='relu'))
model.add(Conv2D(filters = 32,kernel_size = (3,3),padding = 'same',activation='relu'))
model.add(MaxPool2D(pool_size=(2,2),strides = (2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(256,activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(10, activation = "softmax"))

```

```python
model.summary()
```

```python
Optimizer = RMSprop(learning_rate=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
```

```python
model.compile(optimizer = Optimizer , loss = "categorical_crossentropy", metrics=["accuracy"])
```

```python
from re import X
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
        featurewise_center=False,  # set input mean to 0 over the dataset
        samplewise_center=False,  # set each sample mean to 0
        featurewise_std_normalization=False,  # divide inputs by std of the dataset
        samplewise_std_normalization=False,  # divide each
        zca_whitening=False,  # apply ZCA whitening
        rotation_range=10,  # randomly rotate images in the range (degrees, 0 to 180)
        zoom_range = 0.1, # Randomly zoom image
        width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width)
        height_shift_range=0.1,  # randomly shift images vertically (fraction of total height)
        horizontal_flip=False,
        vertical_flip=False)
datagen.fit(X_train)
```

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
learning_rate_reduction = ReduceLROnPlateau(monitor = 'val_acc',patience = 3,verbose = 1,factor = 0.5,min_lr = 0.00001)
```
